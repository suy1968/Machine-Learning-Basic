## Part 1

```python
import pandas as pd
import numpy as np

dataset = pd.read_csv('train.csv')

# 1 Dropping a Single Column of a dataset based on the column name
df1 = dataset.drop('Name',axis=1)
df1.head()
```

```python
# 2.Dropping multiple Column of a dataset based on the column name

df1 = df1.drop(['Pclass','Cabin'],axis=1)
df1.head()
```

```python
# 3. Dropping the Column of a dataset based on the Column Index
df1 = df1.drop(df1.columns[1],axis=1)
df1.head()
```

```python
# 4. Dropping the Multiple Column of a dataset based on the Column Index
df2 = df1.drop(df1.columns[[1,3]],axis=1)
df2.head()
```

```python
# 5. Deleting rows of a dataset and copying the resultant dataset to a new dataframe
df3 = df2.drop(df2.index[[1,3]])
df3.head()
```

```python
# 6. Drop the Duplicate Values based on some conditions
df3 = df3[df3['Embarked'] != 'S'].head()
df3.head()
```

```python
Dup = df3.drop_duplicates().head()
Dup = Dup.drop_duplicates(subset=['Embarked'])
Dup
```

```python
Dup1 = df3.drop_duplicates(subset=['Embarked'], keep='last')
Dup1
```

## Part 3

```python
#suppose the date range start from 1st Jan, 2017 and there are 200000 entires
# with the frequency of every 1 second
ts_index = pd.date_range('01/01/2017',periods=200000,freq='60S')

# Create the dataframe
df = pd.DataFrame(index=ts_index)

# We add a column 'No_of_Vehicles' and generate a random number between 1 to 10
df['No_of_Vehicles'] = np.random.randint(1,10,200000)
```

```python
# Groupby on a weekly basis
df.resample('W').sum()
```

```python
# Groupby based on bi-weekly basis
df.resample('2W').sum()
```

```python
# Groupby on a monthly basis
df.resample('M').sum()
```

```python
df.resample('M',label='left').sum()
```

## PART 4 

```python
ts_value = pd.date_range('01/01/2017',periods=200000,freq='15S')

# Create the dataframe df
df = pd.DataFrame()

# Add a column with ts_value above
df['Datetime'] = ts_value

# Dervie a column year from the Column "Datetime" 
df['Year'] = df.Datetime.dt.year

# Dervie a column month from the Column "Datetime"
df['Month'] = df.Datetime.dt.month

# Dervie a column day from the Column "Datetime"
df['Day'] = df.Datetime.dt.day
```

```python
df.head()
```

```python
# Dervie a column week_day from the Column "Datetime"
df['Weekday'] = df.Datetime.dt.weekday

# Dervie a column Hour from the Column "Datetime"
df['Hour'] = df.Datetime.dt.hour

# Dervie a column minutes from the Column "Datetime"
df['Minutes'] = df.Datetime.dt.minute

# Dervie a column second from the Column "Datetime"
df['Second'] = df.Datetime.dt.second

# Dervie a column microsecond from the Column "Datetime"
df['Microsecond'] = df.Datetime.dt.microsecond
```
