## Part - 1


```python
import pandas as pd

train = pd.read_csv('train.csv')
```

```python
df = pd.DataFrame()

df['Name'] = train['Name']
df['Pclass'] = train['Pclass']
```

```python
df.sort_values(by='Name',ascending=0,inplace=True)
df
```

```python
df.sort_values(by='Name',ascending=1,inplace=True)
df
```

```python
df.sort_values(by=['Name','Pclass'],ascending=[True,False],inplace=True)
df
```

```python
df1 = df.sort_values(by=['Pclass','Name'],ascending=[True,True])
print(df1)
df2 = df.sort_values(by=['Pclass','Name'],ascending=[True,False])
print(df2)
df3 = df.sort_values(by=['Pclass','Name'],ascending=[False,True])
print(df3)
df4 = df.sort_values(by=['Pclass','Name'],ascending=[False,False])
print(df4)
```
