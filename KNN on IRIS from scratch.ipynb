```
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets
import numpy as np
```

```
iris = datasets.load_iris()
```

```
print('data shape = {}'.format(iris.data.shape))
print('target shape = {}'.format(iris.target.shape))
```

```
X = iris.data[:, :2]
y= iris.target
print(y)
x=X[-1]
print('------------------------')
print(X.shape)
print('------------------------')
print(y.shape)
print('------------------------')
print(x)
x= x.reshape(1,-1)
print('------------------------')
print(x)
print('------------------------')
print(x.shape)
```

```
N,D = X.shape
M,D = x.shape
```

```
k=5
dist = np.zeros((N,M))
for i in range(N):
  for j in range(M):
    dist[i,j] = np.sqrt((X[i]-x[j]).T.dot(X[i]-x[j]))

print('Distance Matrix = {}'.format(dist))
```

```
num_classes = len(np.unique(y))
ypred = np.zeros(num_classes)
classes = y[np.argsort(dist,axis=0)][:k]  #finds labels of nearest neighbours
print('K nearest classes = ',classes)
```

```
for c in np.unique(classes):
  ypred[c] = (classes == c).sum()

print('ypred = ',ypred)
print('max = ',ypred.max())
yp = np.argmax(ypred)
print("Predicted class for x is : ",yp," i.e ",iris.target_names[yp] )
```
